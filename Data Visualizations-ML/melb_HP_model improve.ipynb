```python
from sklearn.model_selection import train_test_split
x_tr,x_tes,y_tr,y_tes = train_test_split (x,y,test_size=0.25, random_state =0)
```

```python

# Define model. Specify a number for random_state to ensure same results each run
model = RandomForestRegressor(random_state=1)
#RandomForestRegressor
# Fit model:      DecisionTreeRegressor
model.fit(x_tr, y_tr)
#coeff_df = pd.DataFrame(model.coef_,x.columns,columns = ['Coefficient'])
```

```python
y_pred=model.predict(x_tes)
print(y_pred)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_tes, y_pred))  
print('R2:', metrics.r2_score(y_tes,y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_tes, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_tes, y_pred))) 
```

```python
def get_mae(max_leaf_nodes,x_tr, x_tes,y_tr, y_tes):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(x_tr,y_tr)
    preds_val = model.predict(x_tes)
    mae = metrics.mean_absolute_error(y_tes, preds_val)
    r2=metrics.r2_score(y_tes,preds_val)
    return (mae,r2)
for max_leaf_nodes in [5,10,100,250,500,1000,5000]:
    my_mae, R2= get_mae(max_leaf_nodes, x_tr, x_tes, y_tr, y_tes)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d \t\t  R2:  %f" %(max_leaf_nodes, my_mae,R2))
```
